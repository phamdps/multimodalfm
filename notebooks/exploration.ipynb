## Why this is happening

Jupyter's Current Working Directory (CWD): When you run a Jupyter Notebook, the "current directory" is not always the folder where the .ipynb file is saved [1.3.1]. It is usually the directory where you launched the Jupyter server (e.g., your home folder or a parent folder) [1.3.3].

Relative Paths: If your code says open("models/file.pt", "w"), it looks for a models/ folder inside the directory where the Jupyter server started, not necessarily inside your project folder [1.3.1].

Visual Filters: If you are using VS Code or another IDE, check that you don't have "hidden files" or "excluded files" settings that are making your new files invisible in the sidebar [1.4.2].
